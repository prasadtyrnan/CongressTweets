```python
import pandas as pd
import numpy as np
```

```python
tweets_df = pd.read_csv("Tweets11122022.csv")
tweets_df = tweets_df.drop(3892).reset_index()
tweets_df.head()
```

```python
#First I want to create a dataset of nodes for the tweets
tweet_nodes = pd.DataFrame({"id:ID(Tweet)" : np.array(tweets_df.index), "Content" : np.array(tweets_df.Content), "Time" : np.array(tweets_df.Time), ":LABEL" : np.repeat("Tweet", len(tweets_df.index))})
tweet_nodes.to_csv("nodes/tweets.csv", index = False)
```

```python
tweet_nodes
```

```python
#Now I want to create relationships between each tweet and the tweeter
tweet_id = np.array(tweets_df.index)
tweeter = np.array(tweets_df.Name)

tweet_relationships = pd.DataFrame({":START_ID(Congresspeople-ID)" : tweeter, ":END_ID(Tweet)" : tweet_id, ":TYPE" : np.repeat("tweeted", len(tweet_id))})
tweet_relationships.to_csv("relationships/congressperson_to_tweet.csv", index = False)
tweet_relationships.head()
```

```python
#Now I'm going to create edges from each tweet to its hashtags
hashtags = []
origin_tweet = []
for i in range(len(tweets_df.index)):
    hashtag = "".join(str(tweets_df['Hashtags'][i])[1:-1].split("'")).split(", ")
    for h in hashtag:
        if h != '':
            hashtags.append(h)
            origin_tweet.append(i)
            
hashtags_relationships = pd.DataFrame({":START_ID(Tweet)" : origin_tweet, ":END_ID(Hashtag)" : hashtags, ":TYPE" : np.repeat("has hashtag", len(hashtags))})
hashtag_df = pd.DataFrame({"hashtag:ID(Hashtag)" : np.unique(hashtags), ":LABEL" : np.repeat("Hashtag", len(np.unique(hashtags)))})
hashtags_relationships.to_csv("relationships/hashtag_relationships.csv", index = False)
hashtag_df.to_csv("nodes/hashtags.csv", index = False)
hashtags_relationships.head()
```

```python
#Now I'm going to extract the annotations
annotations = []
annotation_type = []
annotation_target = []
origin_tweet = []
for i in range(len(tweets_df.index)):
    annotation = "".join(str(tweets_df['Annotations'][i])[1:-1].split("'")).split(", ")
    for a in annotation:
        if a != '':
            try:
                annotations.append(a)
                annotation_type.append(a.split(":")[0].replace('"', ""))
                annotation_target.append(a.split(":")[1].replace('"', ""))
                origin_tweet.append(i)
            except Exception as e:
                print(e, a, i)

annotations = np.array(annotations)
annotation_type = np.array(annotation_type)
annotation_target = np.array(annotation_target)
origin_tweet = np.array(origin_tweet)

annotation_types = np.unique(annotation_type)

for at in annotation_types:
    targets_filtered = np.array(np.unique(annotation_target[annotation_type == at]))
    targets_df = pd.DataFrame({"name:ID(" + at + ")" : targets_filtered, ":LABEL" : np.repeat(at, len(targets_filtered))})
    targets_df.to_csv("nodes/" + at.lower() + ".csv", index = False)
                      
    targets_start = origin_tweet[annotation_type == at]
    targets_end = annotation_target[annotation_type == at]
    relationships_filtered = pd.DataFrame({":START_ID(Tweet)" : targets_start, ":END_ID(" + at + ")" : targets_end, ":TYPE" : np.repeat("references", len(targets_start))})
    relationships_filtered.to_csv("relationships/" + at.lower() + "_relationships.csv", index = False)
```

```python
origin_tweet = []
types = []
targets = []
specifications = []
for i in range(len(tweets_df.index)):
    cannotation = tweets_df['Context Annotations'][i][1:-1].split("',")
    for ca in cannotation:
        if ca[:1] == "'" or ca[:1] == '"':
            ca = ca[1:]
        elif ca[:2] == " '" or ca[:2] == ' "':
            ca = ca[2:]
        if ca[:1] == '"':
            ca = ca[1:]
            
        if ca[-1:] == '"' or ca[-1:] == "'":
            ca = ca[:-1]
        ca = ca.split(",")
        if len(ca) < 2:
            print(ca, end = "\r")
        else:
            origin_tweet.append(i)
            types.append(ca[0])
            targets.append(ca[1])
            if len(ca) > 2:
                specifications.append(",".join(ca[2:]))
            else:
                specifications.append('')
                
origin_tweet = np.array(origin_tweet)
types = np.array(types)
targets = np.array(targets)
specifications = np.array(specifications)

unique_types = np.unique(types)
for ut in unique_types:
    idx = np.where(types == ut)
    
    targets_filtered = targets[idx]
    origins_filtered = origin_tweet[idx]
    specifications_filtered = specifications[idx]
    
    targets_df = pd.DataFrame({"name:ID(" + ut + "-ca)" : np.unique(targets_filtered), ":LABEL" : np.repeat(ut, len(np.unique(targets_filtered)))})
    relationships_filtered = pd.DataFrame({":START_ID(Tweet)" : origins_filtered, ":END_ID(" + ut + "-ca)" : targets_filtered, ":TYPE" : np.repeat("references", len(targets_filtered)), "Specification" : specifications_filtered})
    
    targets_df.to_csv("nodes/" + ut.lower() + "_ca.csv", index = False)
    relationships_filtered.to_csv("relationships/" + ut.lower() + "relationships_ca.csv", index = False)
```
