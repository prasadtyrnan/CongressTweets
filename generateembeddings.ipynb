```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
```

```python
date = "02052022"
```

```python
df = pd.read_csv("Tweets" + date + ".csv")
df = df.loc[df['Retweeting'] == "$"]
df.index = np.arange(0, len(df))
df
```

```python
model = SentenceTransformer('distilbert-base-nli-mean-tokens')
```

```python
data = df["Content"]
h_embs = []
c_embs = []
for i in range(len(data)):
    print(100*((i+1)/len(data)), end = "\r")
    c_embs.append(model.encode(data[i], show_progress_bar = False))
    if df["Hashtags"][i] == "[]":
        h_embs.append(np.zeros((768)))
    else:
        hashtags = df["Hashtags"][i]
        hashtags = hashtags[1:-1]
        h_embs.append(model.encode(df["Hashtags"][i]))
    
h_embs = np.array(h_embs)
c_embs = np.array(c_embs)
```

```python
np.save(date + "_corpus_embeddings", c_embs)
np.save(date + "_hashtag_embeddings", h_embs)
```

```python
hashtags = []
for hashtag in df['Hashtags']:
    hashtag = hashtag[1:-1]
    tags = []
    for tag in hashtag.split("\'"):
        if tag != "" and tag != ", ":
            hashtags.append(tag)
            
hashtags = np.unique(hashtags)
print(len(hashtags))

one_hots = np.zeros((len(df['Hashtags']), len(hashtags)))
for i in range(len(one_hots)):
    print(i / len(one_hots), end = "\r")
    for j in range(len(one_hots[i])):
        if hashtags[j] in df['Hashtags'][i]:
            one_hots[i][j] = 1
            
one_hots
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 32)
oh_pcad = pca.fit_transform(one_hots)
print(pca.explained_variance_)
print(np.sum(pca.explained_variance_))
```
