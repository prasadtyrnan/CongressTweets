```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv("Tweets06112021.csv")
df = df.loc[df['Retweeting'] == "$"]
df.index = np.arange(0, len(df))
df.head()
```

```python
xyz = np.load("6_11_21_topics3dPCA.npy")
cluster_label = np.load("6_11_21_cluster_labels.npy")
cluster_label.shape
```

```python
content_with_br = []
for tweet in df.Content:
    i = 0
    num_chars = 0
    while (i < len(tweet)):
        if tweet[i] == " " and num_chars > 30:
            tweet = tweet[:i] + "<br>" + tweet[i:]
            i += 4
            num_chars = 0
        i += 1
        num_chars += 1
    content_with_br.append(tweet)
```

```python
df["ContentBreaks"] = content_with_br
df.head()
```

```python
df['x'] = xyz[:,0]
df['y'] = xyz[:,1]
df['z'] = xyz[:,2]
df['cluster'] = cluster_label
df.head()
```

```python
fig = px.scatter_3d(df, x = "x", y="y", z="z", hover_name = "Name", color="cluster", hover_data = {"x": False, "y":False, "z":False, "ContentBreaks":True})
fig.update(layout_coloraxis_showscale=False)
fig.update_traces(marker=dict(size=2))
fig.update_layout(
    scene = dict(
        xaxis = dict(visible=False),
        yaxis = dict(visible=False),
        zaxis = dict(visible=False),
    )
)
fig.write_html("test.html")
```
