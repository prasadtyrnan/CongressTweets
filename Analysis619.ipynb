```python
def proportion_in(df_1, df_2, column, value):
    return len(df_1.loc[df_1[column].str.contains(value)]) / len(df_2.loc[df_2[column].str.contains(value)])
```

```python
juneteenth_tweets = pd.merge(tweets.loc[tweets['Hashtags'].str.contains(",Juneteenth,")].sort_values("Time"),cong,how="inner",on="Twitter_Handle")
juneteenth_tweets_d = juneteenth_tweets.loc[juneteenth_tweets["Party_Affiliation"] == "D"]
juneteenth_tweets_r = juneteenth_tweets.loc[juneteenth_tweets["Party_Affiliation"] == "R"]
flagday_tweets = pd.merge(tweets.loc[tweets['Hashtags'].str.contains(",FlagDay,")].sort_values("Time"),cong,how="inner",on="Twitter_Handle")
flagday_tweets_d = flagday_tweets.loc[flagday_tweets["Party_Affiliation"] == "D"]
flagday_tweets_r = flagday_tweets.loc[flagday_tweets["Party_Affiliation"] == "R"]
```

```python
dates = []
rj_count = []
dj_count = []
rf_count = []
df_count = []
for i in range(12,21):
    dates.append("6-" + str(i))
    rj_count.append(len(juneteenth_tweets_r.loc[juneteenth_tweets_r["Time"] <= "2021-06-" + str(i) + "T00:00:00.000Z"]["Time"]))
    dj_count.append(len(juneteenth_tweets_d.loc[juneteenth_tweets_d["Time"] <= "2021-06-" + str(i) + "T00:00:00.000Z"]["Time"]))
    rf_count.append(len(flagday_tweets_r.loc[flagday_tweets_r["Time"] <= "2021-06-" + str(i) + "T00:00:00.000Z"]["Time"]))
    df_count.append(len(flagday_tweets_d.loc[flagday_tweets_d["Time"] <= "2021-06-" + str(i) + "T00:00:00.000Z"]["Time"])) 
```

```python
import matplotlib.pyplot as plt

plt.plot(dates, dj_count, label="Democrats-Juneteenth")
plt.plot(dates, rj_count, label="Republicans-Juneteenth")
plt.plot(dates, df_count, label="Democrats-FlagDay")
plt.plot(dates, rf_count, label="Republicans-FlagDay")
plt.legend()
plt.show()
```
