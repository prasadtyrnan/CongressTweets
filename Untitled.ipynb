```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.congress.gov/congressional-record/2022/01/18/senate-section"
page = requests.get(url)

text = page.text
soup = BeautifulSoup(page.content, "html.parser")
```

```python
links = soup.find_all("a")

new_links = []
for link in links:
    link = (str(link).split("href=\"")[1].split("\"")[0])
    if link[:22] == "/congressional-record/":
        if len(link.split("/")) > 5:
            if link.split("/")[5] == "senate-section":
                new_links.append(link)
links = new_links[1:].copy()
```

```python
big_text = None

for url in links:
    url = "https://www.congress.gov" + url
    print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    txt_links = soup.find_all("a")
    for txt_link in txt_links:
        if "View TXT in new window" in txt_link:
            txt_link = (str(txt_link).split("href=\"")[1].split("\"")[0])
            page = requests.get("https://www.congress.gov" + txt_link)
            
            if big_text:
                big_text += page.text
            else:
                big_text = page.text
```

```python
headers = []
for bt in big_text.split("\n\n"):
    if not "From the Congressional Record Online through the Government Publishing Office [<a href='https://www.gpo.gov'>www.gpo.gov</a>]" in bt:
        if not "<pre>" in bt:
            if bt.lower().islower():
                if bt.upper() == bt:
                    print("Header")
                    headers.append(bt)
                print(bt)
            else:
                print("Ass3")
        else:
            print("Ass2")
    else:
        print("Ass1")
```

```python
section_titles = []
speaker = []
someone_speaking = False
for bt in big_text.split("\n"):
    if len(bt) > 0 and (bt.lower().islower() or True in [char.isdigit() for char in bt]):
        if not "<pre>" in bt and not "[Page S" in bt and not "[Pages S" in bt and not "From the Congressional Record Online through the Government Publishing Office [<a href='https://www.gpo.gov'>www.gpo.gov</a>]" in bt:
            if bt.upper() == bt:
                section_titles.append(bt)
            else:
                if bt[:5] == "  Mr." or bt[:5] == "  Ms." or bt[:6] == "  Mrs.":
                    if bt.split(".")[1][1].isupper() and bt.split(".")[1][-1].isupper():
                        speaker.append(".".join(bt.split(".")[:2]))
                        if someone_speaking:
                            print("End Speaker\n\n")
                        print("\n\nSpeaker:", speaker[-1])
                        someone_speaking = True
                elif bt.split(".")[0] == "  The PRESIDENT pro tempore":
                    if someone_speaking:
                        print("End Speaker\n\n")
                    speaker.append("The PRESIDENT pro tempore")
                    print("\n\nSpeaker:", speaker[-1])
                    someone_speaking = True
                elif bt[:1] == " " and someone_speaking:
                    print("End Speaker\n\n")
                    someone_speaking = False
                    
                print(bt)
```
