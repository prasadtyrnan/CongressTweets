```python
import tensorflow as tf
import tensorflow_probability as tfp
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
data = scipy.io.loadmat('Data/cylinder_nektar_wake.mat')

N_train = 5000

U_star = data['U_star']
P_star = data['p_star']
t_star = data['t']
X_star = data['X_star']

N = X_star.shape[0]
T = t_star.shape[0]

XX = np.tile(X_star[:,0:1], (1,T))
YY = np.tile(X_star[:,1:2], (1,T))
TT = np.tile(t_star, (1,N)).T

UU = U_star[:,0,:]
VV = U_star[:,1,:]
PP = P_star

x = XX.flatten()[:,None]
y = YY.flatten()[:,None]
t = TT.flatten()[:,None]

u = UU.flatten()[:,None]
v = VV.flatten()[:,None]
p = PP.flatten()[:,None]

idx = np.random.choice(N*T, N_train, replace=False)
x_train = x[idx,:]
y_train = y[idx,:]
t_train = t[idx,:]
u_train = u[idx,:]
v_train = v[idx,:]
p_train = p[idx,:]

idx_test = np.random.choice(N*T, 200, replace=False)
x_test = x[idx_test,:]
y_test = y[idx_test,:]
t_test = t[idx_test,:]
u_test = u[idx_test,:]
v_test = v[idx_test,:]
p_test = p[idx_test,:]

X_train = tf.cast(tf.concat([x_train, y_train, t_train], 1), dtype=tf.float32)
Y_train = tf.cast(tf.concat([u_train, v_train], 1), dtype=tf.float32)

X_test = tf.cast(tf.concat([x_test, y_test, t_test], 1), dtype=tf.float32)
Y_test = tf.cast(tf.concat([u_test, v_test], 1), dtype=tf.float32)

ub = (X_train).numpy().max(0)
lb = (X_train).numpy().min(0)
```

```python
layers = [3, 20, 20, 20, 20, 20, 20, 2]

params = tf.Variable([], dtype = tf.float32)

def xavier_init(size):
    in_dim = size[0]
    out_dim = size[1]
    xavier_stdev = np.sqrt(2/(in_dim + out_dim))
    return tf.Variable(tf.random.truncated_normal([in_dim*out_dim], stddev=xavier_stdev), dtype = tf.float32)

def initialize_NN(layers):
    num_layers = len(layers)
    weights = tf.Variable([], dtype=tf.float32)
    biases = tf.Variable([], dtype=tf.float32)
    for l in range(len(layers)-1):
        W = xavier_init(size = [layers[l], layers[l+1]])
        b = tf.Variable(tf.zeros([layers[l+1]], dtype=tf.float32), tf.float32)
        weights = tf.concat([weights, W], 0)
        biases = tf.concat([biases, b], 0)
    return tf.concat([weights, biases], 0)

def convert_params(params):
    lower_range = 0
    upper_range = lower_range + layers[0]*layers[1]
    weights = []
    for l in range(1, len(layers)-1):
        weights.append(tf.reshape(params[lower_range:upper_range], [layers[l-1],layers[l]]))
        lower_range = upper_range
        upper_range += layers[l]*layers[l+1]
    weights.append(tf.reshape(params[lower_range:upper_range], [layers[-2], layers[-1]]))
    biases = []
    for l in range(1, len(layers)):
        lower_range = upper_range
        upper_range += layers[l]
        biases.append(tf.reshape(params[lower_range:upper_range],[1,layers[l]]))
    return weights, biases

def predict(weights, biases, inputs):
    num_layers = len(weights) + 1
    H = (2.0*(inputs-lb)/(ub-lb)-1.0)
    H = tf.cast(H, tf.float32)
    for l in range(num_layers-2):
        W = weights[l]
        b = biases[l]
        H = tf.tanh(tf.add(tf.matmul(H,W), b))
    W = weights[-1]
    b = biases[-1]
    return tf.add(tf.matmul(H, W), b)

def loss(params):
    weights, biases = convert_params(params)
    preds = predict(weights, biases, X_train)
    return tf.math.squared_difference(preds[:,0], Y_train[:,0]) + tf.math.squared_difference(preds[:,1], Y_train[:,1])

def reduce_sum_loss(params):
    return tf.reduce_sum(loss(params))

def loss_and_gradient(params):
    return tfp.math.value_and_gradient(reduce_sum_loss, params)

params = initialize_NN(layers)
```

```python
optim_results = tfp.optimizer.bfgs_minimize(loss_and_gradient, initial_position=params, max_iterations=500)
print(optim_results.converged)
```

```python
tr_weights, tr_biases = convert_params(optim_results.position)
preds = predict(tr_weights, tr_biases, X_train)

plt.scatter(Y_train[:,0], preds[:,0], label = "u")
plt.show()

plt.scatter(Y_train[:,1], preds[:,1], label = "v")
plt.show()
```

```python
def loss(params):
    weights, biases = convert_params(params)
    x_tf = tf.reshape(X_train[:,0], [N_train, 1])
    y_tf = tf.reshape(X_train[:,1], [N_train, 1])
    t_tf = tf.reshape(X_train[:,2], [N_train, 1])
    with tf.GradientTape(persistent=True) as tape2:
        tape2.watch(x_tf)
        tape2.watch(y_tf)
        with tf.GradientTape(persistent=True) as tape:
            tape.watch(x_tf)
            tape.watch(y_tf)
            tape.watch(t_tf)
            u_pred = predict(weights, biases, tf.concat([x_tf,y_tf,t_tf], 1))[:,0]
            v_pred = predict(weights, biases, tf.concat([x_tf,y_tf,t_tf], 1))[:,1]
            p_pred = predict(weights, biases, tf.concat([x_tf,y_tf,t_tf], 1))[:,2]
        u_x = tape.gradient(u_pred, x_tf)
        v_x = tape.gradient(v_pred, x_tf)
        u_y = tape.gradient(u_pred, y_tf)
        v_y = tape.gradient(v_pred, y_tf)
        u_t = tape.gradient(u_pred, t_tf)
        v_t = tape.gradient(v_pred, t_tf)
        p_x = tape.gradient(p_pred, x_tf)
        p_y = tape.gradient(p_pred, y_tf)
    u_xx = tape2.gradient(u_x, x_tf)
    u_yy = tape2.gradient(u_y, y_tf)
    v_xx = tape2.gradient(v_x, x_tf)
    v_yy = tape2.gradient(v_y, y_tf)
    
    preds = predict(weights, biases, tf.concat([x_tf, y_tf, t_tf], 1))
    u_pred = preds[:,0]
    v_pred = preds[:,1]
    
    f_u = u_t + (u_pred*u_x + v_pred*u_y) + p_x + 0.01*(u_xx + u_yy)
    f_v = v_t + (u_pred*v_x + v_pred*v_y) + p_y + 0.01*(v_xx + v_yy)
    
    return tf.math.squared_difference(u_pred, Y_train[:,0]) + tf.math.squared_difference(v_pred, Y_train[:,1]) + (f_u ** 2) + (f_v ** 2)

def reduce_sum_loss(params):
    loss_tr = loss(params)
    print(tf.reduce_mean(loss_tr))
    return tf.reduce_sum(loss_tr)

def loss_and_gradient(params):
    return tfp.math.value_and_gradient(reduce_sum_loss, params)

optim_results = tfp.optimizer.lbfgs_minimize(loss_and_gradient, initial_position=params, max_iterations=1000)
print(optim_results.converged)
```
