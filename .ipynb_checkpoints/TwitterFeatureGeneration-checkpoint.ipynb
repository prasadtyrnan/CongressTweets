```python
import pandas as pd

tweets = pd.read_csv("Tweets06112021.csv")
tweets.head()
```

```python
import re
Retweeting = []
Tagged = []
Hashtags = []
Links = []

for i in range(len(tweets['Name'])):
    Retweet = ''
    Tags = ''
    Hashtag = ''
    Link = ''
    content = tweets['Content'][i]
    print(str(i) + ": " + content)
    if content[:2] == 'RT':
        #print("Retweeting: " + content[4:].split(':')[0])
        Retweet += content[4:].split(':')[0] + ", "
    else:
        for j in range(len(content)):
            if content[j] == "@":
                    #print(re.sub('\W+', '', re.split(' |\n', content[j:])[0]))
                Tags += re.sub('\W+', '', re.split(' |\n', content[j:])[0]) + ", "
            elif content[j] == "#":
                #print(re.sub('\W+', '', re.split(' |\n', content[j:])[0]))
                Hashtag += re.sub('\W+', '', re.split(' |\n', content[j:])[0]) + ", "
        words = re.split(' |\n', content)
        for word in words:
            if (len(word) > 4 and word[:4] == 'http'):
                Link += word + ', '
    if len(Retweet) > 0:
        Retweet = Retweet[0:-2]
    else:
        Retweet = "$"
    if len(Tags) > 0:
        Tags = Tags[0:-2]
    else:
        Tags = "$"
    if len(Hashtag) > 0:
        Hashtag = Hashtag[0:-2]
    else:
        Hashtag = "$"
    if len(Link) > 0:
        Link = Link[0:-2]
    else:
        Link = "$"
    Retweeting.append(Retweet)
    Tagged.append(Tags)
    Hashtags.append(Hashtag)
    Links.append(Link)
```

```python
tweets['Retweeting'] = Retweeting
tweets['Hashtags'] = Hashtags
tweets['Tagged'] = Tagged
tweets['Links'] = Links
tweets.head()
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

tokenizer = AutoTokenizer.from_pretrained("dslim/bert-base-NER")
model = AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)

example = tweets['Content'][12]
print(example)

ner_results = nlp(example)
print(ner_results)
```
