```python
import pandas as pd
df = pd.read_csv("CongressData.csv")
#Load data for congresspeople
df.head()
```

```python
#Used to allow different strings to be used as keys while retaining legibility (ex: Jesús García -> JesusGarcia)
def replace_special_chars(string):
    return string.replace(" ", "").replace("ú","u").replace("í","i").replace("á","a").replace("é", "e").replace("ü", "u").replace("ñ", "n")
```

```python
#Sign in to local arango database, would need to be changed if it isn't being hosted on local machine
from pyArango.connection import *

conn = Connection(username="root", password="")
db = conn["Congress"]
db
```

```python
#Creates and populates a collection of documents for each Congressperson
Congresspeople = db.createCollection(name="Congresspeople")
for i in range(len(df['Congressperson'])):
    try:
        dude = df["Congressperson"][i]
        doc = Congresspeople.createDocument()
        doc._key = replace_special_chars(dude)
        doc["Name"] = dude
        doc["District"] = df["State"][i] + "-" + df["District"][i]
        doc["Birthdate"] = df["Birthdate"][i]
        doc["Twitter Handle"] = df["Twitter Handle"][i]
        doc.save()
    except:
        print("Something went wrong with {}".format(dude))
```

```python
Party = db.createCollection(name="Parties")
Republicans = Party.createDocument()
Republicans._key = "R"
Democrats = Party.createDocument()
Democrats._key = "D"
Independents = Party.createDocument()
Independents._key = "I"
Republicans.save()
Democrats.save()
Independents.save()
```

```python
Chambers = db.createCollection(name="Chambers")
House = Chambers.createDocument()
House._key = "House"
House.save()
Senate = Chambers.createDocument()
Senate._key = "Senate"
Senate.save()
```

```python
State = db.createCollection(name="States")
for s in df["State"].unique():
    state = State.createDocument()
    state._key = s
    state.save()
```

```python
Race = db.createCollection(name="Races")
for race in df["Race"].unique():
    try:
        racey = Race.createDocument()
        racey._key = str(race.split(",")[0])
        racey.save()
    except:
        print("Something went wrong with {}".format(race))
```

```python
Background = db.createCollection(name="Backgrounds")
for background in df["Background"].unique():
    try:
        back = Background.createDocument()
        back._key = str(background.split(",")[0])
        back.save()
    except:
        print(background.split(",")[0])
```

```python
Edges = db.createCollection(name="Edges", className='Edges')
for i in range(len(df["State"])):
    edge = Edges.createEdge()
    edge["Type"] = "From State"
    edge._from = "Congresspeople/" + df["Congressperson"][i].replace(" ", "").replace("ú","u").replace("í","i").replace("á","a").replace("é", "e")
    edge._to = "States/" + df["State"][i]
    edge.save()
```

```python
for i in range(len(df["Party Affiliation"])):
    edge = Edges.createEdge()
    edge["Type"] = "Affiliation"
    edge._from = "Congresspeople/" + replace_special_chars(df["Congressperson"][i])
    edge._to = "Parties/" + df["Party Affiliation"][i]
    edge.save()
```

```python
for i in range(len(df["House or Senate"])):
    edge = Edges.createEdge()
    edge["Type"] = "Which Chamber"
    edge._from = "Congresspeople/" + replace_special_chars(df["Congressperson"][i])
    edge._to = "Chambers/" + df["House or Senate"][i]
    edge.save()
```

```python
for i in range(len(df["Background"])):
    back = df["Background"][i]
    for bac in back.split(","):
        edge = Edges.createEdge()
        edge._from = "Congresspeople/" + replace_special_chars(df["Congressperson"][i])
        edge._to = "Backgrounds/" + bac.replace(" ", "")
        edge["Type"] = "Background"
        edge.save()
```

```python
for i in range(len(df["Race"])):
    races = df["Race"][i]
    for race in races.split(","):
        edge = Edges.createEdge()
        edge._from = "Congresspeople/" + replace_special_chars(df["Congressperson"][i])
        edge._to = "Races/" + race.replace(" ", "")
        edge["Type"] = "Of Race"
        edge.save()
```
